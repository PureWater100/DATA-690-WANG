# Brief description: 
# The purpose of this program is to take 10 integers given by the user and produce some basic statistical calculations. 
## What will be displayed in order:
## - The 10 integers the user entered
## - Minimum
## - Maximum
## - Range
## - Variance
## - Standard Deviation
## Note:
## There are no packages or statistical functions used. Instead the statistical calculations were done through the use of list sorting and for loops. 

## Accepting User Input

```
user_inputs = []
MAX_TRY = 11

for i in range(1, MAX_TRY):

    while True:
        try:
            user_input = input("Please enter an integer:")
            in_input = int(user_input)
            break
        except:
            print("Please retry (only integer accepted)")
    user_inputs.append(user_input)   
    print(f"You have entered integer #{i}:", user_input)

ints = [int(item) for item in user_inputs] #cast each element into an integer 

print("Overall, you have entered:", ints) #display looks cleaner after casting
```

## Doing Statistics: Manipulating the list "ints"

### Sort ints in descending order: 
### - Key to accessing the mininum and maximum
### - Key to calculating the range

```
ints.sort(reverse = True)

# print(ints) will show this list has been sorted in descending order

```

### Access the minimum and maximum 

```
print("The minimum is:", ints[9])
print("The maximum is:", ints[0])
```

### Calculate the Range 

```
the_max = ints[0] 
the_min = ints[9]
print("The range is:", the_max - the_min)
```

### Calculate the Mean:
### - sum up elements in list by using a for loop
### - assign this to sum, and divide sum by 10

```
# Python program to find sum of elements in list
sum = 0 
# Iterate each element in list
# and add them in variale sum
for ele in range(0, len(ints)):
    sum += ints[ele]
mean = sum / 10
print("The mean is:", mean)
```

### Avoiding math/stat functions: How to Calculate the Variance and Standard Deviation 
### - Get the variance by using a for loop
### - Get the standard deviation by raising the variance to the .5 power



```
sum_diff = 0
for ele in range(0, len(ints)):
    sum_diff += (ints[ele] - mean) ** 2
var = sum_diff / 9

# Format variance to display only 2 decimals
variance = "The variance is: {:.2f}"
print(variance.format(var))

# Get standard deviation by raising variance to .5 power
stan_dev = var ** .5

# Format standard deviation to 2 decimals 
stand_dev = "The standard deviation is: {:.2f}"
print(stand_dev.format(stan_dev))
```
