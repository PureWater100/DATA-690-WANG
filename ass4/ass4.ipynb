## - Put the rest of the lines in a new list named "data_list"

```
data_list = []
with open('census_cost.txt', 'rt') as f:
    data_list = f.readlines()
    data_list = data_list[2:] #eliminated first two lines of the file 

for row in data_list:
    print(row)
```

## - Extract the "Total Population" column from the data_list and assign them to a list named "pop_list"
## - Remove the "," from the numbers since Python doesn't recognize them
## - Print the cleansed "pop_list", which is assigned to new_pop_list

### Extracting the second column and assigning it to a new list, pop_list

```
pop_list = []
pop_list = [row[4:17] for row in data_list]



for element in pop_list:
    print(element)

print(pop_list)


```

```

with open("another_new_file.txt", "wt") as f:
    for line in data_list[2:12]:
        f.write(line.replace("\t", "***"))
```

```
def cleanse_pop(pop):

    pop = pop.replace(",", "")
    pop = pop.replace("$","")
    pop = pop.replace("\t","")
    pop = pop.replace("*", "")
    #pop = int(pop)
    return pop
```

### Cleaning pop_list and assigning it to new_pop_list 

```
new_pop_list = [cleanse_pop(pop) for pop in pop_list]

print(new_pop_list)

for element in new_pop_list:
    print(element)
```
