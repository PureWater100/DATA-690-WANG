# 1. Printing a 10 by 10 Matrix of Random Numbers 

```
import random

for i in range(0,10):
    randomlist = []
    for i in range(0,10):
        n = random.randint(0,9)
        randomlist.append(n)
    for x in range(len(randomlist)): 
        print(randomlist[x], end = " ")
    print()
```

### Replace Odd Numbers with @

```
import random
randomlist = []
for i in range(0,10):
    n = random.randint(0,9)
    if (n % 2 != 0):
        randomlist.append("@")
    else:
        randomlist.append(n)

for x in range(len(randomlist)): 
    print(randomlist[x], end = " ")
```

# 2. Replacing Odd Numbers With @

```
import random

for x in range(0, 10):
    randomlist = []
    for i in range(0,10):
        n = random.randint(0,9)
        if (n % 2 != 0):
            randomlist.append("@")
        else:
            randomlist.append(n)

    for x in range(len(randomlist)): 
        print(randomlist[x], end = " ")
    
    print()
```

### Sum of 10 Numbers

```
#add the 10 numbers to get a total and print the 10 numbers followed by "*" and the total.
import random
randomlist = []
for i in range(0,10):
    n = random.randint(0,9)
    randomlist.append(n)
for x in range(len(randomlist)): 
    print(randomlist[x], end = " ")

sum = 0
for ele in range(0, len(randomlist)):
    sum += randomlist[ele]
 
# printing total value
print("*", sum)
```

# 3. Sum of Each Row of Numbers

```
import random

for x in range(0, 10):
    randomlist = []
    for i in range(0,10):
        n = random.randint(0,9)
        randomlist.append(n)
    for x in range(len(randomlist)): 
        print(randomlist[x], end = " ")

    sum = 0
    for ele in range(0, len(randomlist)):
        sum += randomlist[ele]
 
    # printing total value
    print("*", sum)
```
